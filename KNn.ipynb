```python
import numpy as np
import pandas as pd
import math
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset('iris')

print('\n Pierwsze 10 rekordów: ')
print(df.head(10))

print('\n 10 ostatnich rekordów: ')
print(df.tail(10))

print(df.sample(10))

print('\n Liczba próbek w zbiorze: ', len(df))
print("Liczba cech: ", len(df.columns)-1)
print('Nazwy cech: ', [col for col in df.columns if col != "species"])

print('Gatunki: ', df['species'].unique())
print(df['species'].value_counts())
```

```python
sns.histplot(df['species'])
```

```python
feature_pairs = [
    ('sepal_length', 'sepal_width'),
    ('sepal_length', 'petal_length'),
    ('sepal_length', 'petal_width'),
    ('sepal_width', 'petal_width'),
    ('sepal_width', 'petal_length'),
    ('petal_length', 'petal_width')
]

features_names_pl = {
    "sepal_length": "Długość działki kielicha (cm)",
    "sepal_width": "Szerokość działki kielicha (cm)",
    "petal_length": "Długość płatka (cm)",
    "petal_width": "Szerokość płatka (cm)"
}

for x_features, y_features in feature_pairs:
  plt.figure(figsize=(8,6))
  scatter = sns.scatterplot(
      data = df,
      x = x_features,
      y = y_features,
      hue = 'species',
      alpha = 0.8,
      s=100,
      edgecolor = 'k',
      linewidth = 0.5
  )

  x_name_pl = features_names_pl[x_features]
  y_name_pl = features_names_pl[y_features]

  plt.xlabel(x_name_pl, fontsize = 12)
  plt.ylabel(y_name_pl, fontsize = 12)

  plt.legend(title='Gatunek', fontsize = 10)
  plt.grid(True)
  plt.show()
```

```python
import plotly.express as px

px.scatter_3d(x= df['petal_length'], y= df['petal_width'], z = df['sepal_length'], color = df['species'])
```

```python
x = df.drop(columns = 'species').values
y = df['species'].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle= True, stratify = y)

print(f"Liczba próbek w zbiorze uczącym: {x_train.shape[0]}")
print(f"Liczba próbek w zbiorze testowym {x_test.shape[0]}")

unique_train, counts_train = np.unique(y_train, return_counts= True)
for label, count in zip(unique_train, counts_train):
  print(f'{label}: {count} próbek')

unique_test, counts_test = np.unique(y_test, return_counts= True)
for label, count in zip(unique_test, counts_test):
  print(f'{label}: {count} próbek')
```
