```python
import matplotlib.pyplot as plt

dziedzina = [-3, -2, -1, 0, 1, 2, 3]
wynik = [x**2 for x in dziedzina]  # Funkcja kwadratowa

plt.plot(dziedzina, wynik, marker='o', color='red')  # Zmieniamy kolor na czerwony
plt.xlabel("x")
plt.ylabel("g(x)")
plt.title("Wykres funkcji kwadratowej g(x) = x^2 + 2x - 6")
plt.grid()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ustawienia tablicy
size = 5  # Rozmiar tablicy 5x5

# Generowanie losowych punktw
ilosc_punktow = np.random.randint(5, 10)  # Losowa liczba punktw
x = np.random.randint(0, size, ilosc_punktow)  # Losowe wsprzdne x
y = np.random.randint(0, size, ilosc_punktow)  # Losowe wsprzdne y

# Sortujemy punkty wedug wsprzdnej x, aby linia nie skakaa chaotycznie
sorted_points = sorted(zip(x, y))
x_sorted, y_sorted = zip(*sorted_points)

# Tworzenie wykresu
fig, ax = plt.subplots()
ax.set_xticks(range(size))
ax.set_yticks(range(size))
ax.grid(color="black", linestyle="-", linewidth=1)
ax.set_xlim(-0.5, size - 0.5)
ax.set_ylim(-0.5, size - 0.5)
ax.set_title("Tablica 5x5 z poczonymi kropkami")

# Rysowanie punktw jako mae kropki
ax.scatter(x, y, color="red", s=100, zorder=2)  # 's' okrela rozmiar kropek

# Rysowanie linii czcej punkty
ax.plot(x_sorted, y_sorted, color="blue", linewidth=2, zorder=1)

plt.gca().invert_yaxis()  # Odwracamy o Y, by wygldao jak macierz
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generowanie losowych ocen dla 25 uczniw (oceny od 1 do 6)
np.random.seed(42)  # Ustawienie ziarna losowoci dla powtarzalnoci wynikw
oceny = np.random.randint(1, 7, 25)  # Losowe oceny w zakresie 1-6

# Tworzenie histogramu
plt.hist(oceny, bins=np.arange(1, 8) - 0.5, edgecolor="black", color="skyblue")

# Ustawienia wykresu
plt.xticks(range(1, 7))  # Ustawienie etykiet na osi X (1-6)
plt.xlabel("Oceny")
plt.ylabel("Liczba uczniw")
plt.title("Histogram ocen z matematyki")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Wywietlenie wykresu
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generowanie losowych ocen dla 25 uczniw (oceny od 1 do 6)
np.random.seed(42)  # Ustawienie ziarna losowoci dla powtarzalnoci wynikw
oceny = np.random.randint(1, 7, 25)  # Losowe oceny w zakresie 1-6

# Zliczenie liczby wystpie poszczeglnych ocen
unique, counts = np.unique(oceny, return_counts=True)

# Tworzenie diagramu koowego
plt.figure(figsize=(6, 6))
plt.pie(counts, labels=unique, autopct='%1.1f%%', colors=plt.cm.Paired.colors, startangle=90)

# Ustawienia wykresu
plt.title("Diagram koowy ocen z matematyki")

# Wywietlenie wykresu
plt.show()
```
